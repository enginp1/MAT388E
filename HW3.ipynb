```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.io import loadmat
from sklearn.model_selection import cross_val_score
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC 
from sklearn.linear_model import LogisticRegression
from sklearn.cluster import AgglomerativeClustering, DBSCAN
from sklearn import preprocessing
import scipy.cluster.hierarchy as hc
```

```python
pine = loadmat('C:\\Users\\Engin\\Desktop\/data/Indian_pines_corrected.mat')['indian_pines_corrected']
pine_gt = loadmat('C:\\Users\\Engin\\Desktop\/data/Indian_pines_gt.mat')['indian_pines_gt']
X = pine.reshape((145*145,200))
y = pine_gt.reshape(145*145)
```

```python
lb = preprocessing.LabelBinarizer()
y_=lb.fit_transform(y)
```

```python
def model_(model,X,y_,n):
    if model=='LR':
        sc={}
        classifier = LogisticRegression()
        for i in range(y_.transpose().shape[0]):
            scores = cross_val_score(classifier, X, y_.transpose()[i], cv=10)
            sc[i]=scores
        return sc
    elif model=='SVM':
        sc={}
        classifier = SVC()
        for i in range(y_.transpose().shape[0]):
            scores = cross_val_score(classifier, X, y_.transpose()[i], cv=10)
            sc[i]=scores
        return sc
    elif model=='KNN':
        sc={}
        classifier = KNeighborsClassifier(n_neighbors=n)
        for i in range(y_.transpose().shape[0]):
            scores = cross_val_score(classifier, X, y_.transpose()[i], cv=10)
            sc[i]=scores
        return sc
```

```python
LR=model_('LR',X,y_,5)
```

```python
SVM=model_('SVM',X,y_,5)
```

```python
lst={}
for n in range(1,15):
    top=0
    sc=model_('KNN',X,y_,n)
    for i in range(17):
        top=sc[i].mean()+top
    m=top/17
    lst[n]=m
    
    
    
```
